```python
# 1. Definir el diccionario perfil con DATOS REALES (mínimo 5 claves, una con lista)
perfil = {"nombre": "Arvey Suarez","edad": 32,"ubicacion": "Bogotá","transporte": "Moto","postres_favoritos": ["Tres leches", "Milo", "Maracuyá"]
}
```

```python
# 2. Imprimir el menú de opciones (Ejecución única por corrida)
print("--- MENÚ DE GESTIÓN DE PERFIL ---")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con .pop()")
print("5. Consultar un dato con .get()")
print("6. Ver el perfil completo")
print("---------------------------------")
```

```python
# 3. Pedir al usuario que elija una opción con input()
opcion = input("Elige una opción (1-6): ")
```

```python
if opcion == "1":
    # 1. Agregar un dato nuevo (Asignación directa)
    clave = input("Ingresa la nueva clave: ")
    valor = input("Ingresa el valor para la clave: ")
    perfil[clave] = valor
    print("\n[ÉXITO] Dato agregado correctamente.")
    print("Diccionario actualizado:", perfil)
elif opcion == "2":
    # 2. Actualizar un dato existente (Verificación previa con 'in')
    clave = input("Ingresa la clave que quieres actualizar: ")
    if clave in perfil:
        nuevo_valor = input(f"Ingresa el nuevo valor para '{clave}': ")
        perfil[clave] = nuevo_valor
        print(f"\n[ÉXITO] Clave '{clave}' actualizada correctamente.")
        print("Diccionario actualizado:", perfil)
    else:
        print(f"\n[ERROR] La clave '{clave}' no existe en el perfil.")
elif opcion == "3":
    # 3. Eliminar un dato con del (Sin conservar valor y controlando el error)
    clave = input("Ingresa la clave que quieres eliminar con 'del': ")
    if clave in perfil:
        del perfil[clave]
        print(f"\n[ÉXITO] Clave '{clave}' eliminada de forma permanente con 'del'.")
        print("Diccionario actualizado:", perfil)
    else:
        print(f"\n[ERROR] La clave '{clave}' no existe, no se pudo eliminar.")
elif opcion == "4":
    # 4. Eliminar un dato con .pop() (Conservando y mostrando el valor eliminado)
    clave = input("Ingresa la clave que quieres eliminar con '.pop()': ")
    if clave in perfil:
        valor_eliminado = perfil.pop(clave)
        print(f"\n[ÉXITO] Clave '{clave}' eliminada.")
        print(f"El valor que se conservó y recuperó es: {valor_eliminado}")
        print("Diccionario actualizado:", perfil)
    else:
        print(f"\n[ERROR] La clave '{clave}' no existe.")
elif opcion == "5":
    # 5. Consultar un dato con .get() (Búsqueda segura sin romper el programa)
    clave = input("Ingresa la clave que quieres consultar: ")
    resultado = perfil.get(clave, "No disponible")
    print(f"\nResultado de la consulta para '{clave}': {resultado}")
elif opcion == "6":
    # 6. Ver el perfil completo
    print("\n--- PERFIL COMPLETO ACTUAL ---")
    print(perfil)
else:
    # Manejar cualquier opción inválida (números fuera de rango o texto)
    print("\n[ERROR] Opción no válida. Por favor, introduce un número del 1 al 6.")
```
