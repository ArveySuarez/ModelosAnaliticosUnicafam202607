```python
peso   = 70
altura = 1.75
imc    = peso / altura ** 2

print(f"IMC: {imc:.1f}")

if imc < 18.5:
    print("Bajo peso")
elif imc < 25:
    print("Normal")
elif imc < 30:
    print("Sobrepeso")
else:
    print("Obesidad")
```

```python
monto = 120000

if monto < 50000:
    porcentaje = 0.0
elif monto <= 99999:
    porcentaje = 0.10
elif monto <= 199999:
    porcentaje = 0.20
else:
    porcentaje = 0.30

descuento = monto * porcentaje
total_a_pagar = monto - descuento

print(f"Descuento aplicado: ${descuento:,.0f} ({porcentaje * 100:.0f}%)")
print(f"Total a pagar: ${total_a_pagar:,.0f}")
```

```python

numeros_prueba = [15, 9, 10, 7]

for n in numeros_prueba:
    if n % 3 == 0 and n % 5 == 0:
        resultado = "FizzBuzz"
    elif n % 3 == 0:
        resultado = "Fizz"
    elif n % 5 == 0:
        resultado = "Buzz"
    else:
        resultado = n

    print(f"n = {n} -> {resultado}")
```

```python

suma_total = 0

for i in range(1, 101):
    suma_total += i
print(f"Suma de todos los números del 1 al 100: {suma_total}")

suma_pares = 0

for i in range(2, 101, 2):
    suma_pares += i

print(f"Suma de los números pares del 1 al 100: {suma_pares}")
```

```python
precios = [12000, 45000, 8500, 67000, 3200]

contador_caros = 0

for i, precio in enumerate(precios, 1):
    print(f"Item {i}: ${precio:,.0f}")

    if precio > 10000:
        contador_caros += 1
print(f"Artículos que cuestan más de $10,000: {contador_caros}")
```

```python
saldo = 1000
retiro = 150
comision = 20
mes = 0

while saldo > 0:
    mes += 1

    saldo = saldo - retiro - comision

    print(f"Mes {mes} | Saldo restante: ${saldo:,.0f}")

print("---")
print(f"El saldo se agotó. Pasaron {mes} meses en total.")
```
