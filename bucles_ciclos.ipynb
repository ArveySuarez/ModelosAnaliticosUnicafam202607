```python
# generar la suma acumulada de todos los numeros del 1 al 5
numero = 1
while numero < 20:
  numero += 1
  print(f"la sumatoria va en {numero}")
```

```python
suma_total = 0
for numero in range(2, 101, 2):
    suma_total += numero
print("La suma total de los números pares del 1 al 100 es:", suma_total)
```
