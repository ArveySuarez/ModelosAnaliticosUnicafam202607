```python
# una lista vacia
mi_lista=[]
# lista con varios elementos
mi_lista=[1,2,3,"4",5.0,True]
print(mi_lista[2])
```

```python
# cambios en la lista
mi_lista[0]="Arvey"
print(mi_lista)
```

```python
# agregar elementos a la lista
mi_lista.append("Suarez")
print(mi_lista)
```

```python
# agregar elementos en pocision especifica
mi_lista.insert(3,"Bogota")
print(mi_lista)
```

```python
# agregar elementos al final de la lista
mi_lista.extend(["Peru","Argentina"])
print(mi_lista)
```

```python
# quitar elementos de la lista
mi_lista.remove("Peru")
print(mi_lista)
```

```python
print("ACTUALIZACION DE DATOS")
valor_buscar=input("Por favor ingerse el nombre para actualizar")
valor_reemplazar=input("Por favor ingrese valor a reemplazar")
mi_lista[mi_lista.index(valor_buscar)]=valor_reemplazar
print("cambio ejecutado. imprimiendo resultado")
print(mi_lista)
```

```python
print("SUPRESION DE DATOS")
valor_eliminar=input("Por favor ingerse el nombre para eliminar")
mi_lista.remove(valor_eliminar)
print("cambio ejecutado")
print(mi_lista)
```

DICCIONARIOS

```python
mi_diccionario={
    "nombre":"Arvey",
    "ciudad":"Bogota",
    "Empresa":"Decameron"
}
```

```python
# acceder a los elementos del diccionario
print(mi_diccionario["nombre"])
```

```python
# modificar un valor en un diccionario
mi_diccionario["nombre"]="Arvey Suarez"
print(mi_diccionario)
```

```python
# añadir un elemento a un diccionario
mi_diccionario["pais"]="Colombia"
print(mi_diccionario)
```

```python
# añadir un elemento a un diccionario
del mi_diccionario["pais"]
print(mi_diccionario)
```
