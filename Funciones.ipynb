```python
def calificacion_notas(nota):
  if nota>=90:
    print("excelente")
  elif nota>=70 and nota <=89:
    print("Bueno")
  elif nota>=50 and nota <=69:
    print("suficiente")
  else:
    print("Insuficiente")

nota=int(input("Ingresa tu nota: "))
calificacion_notas(nota)
```
